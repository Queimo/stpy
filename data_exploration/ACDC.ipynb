```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "./data/ACDC/full_dataset_exp.csv"
data = pd.read_csv(file_path)

print(data.shape)
print(data.columns)
# Index(['Identifier', 'SMILES', 'Synthesis Procedure', 'Instrument',
#        'Optical Measurements', 'Absorption Wavelength (nm)',
#        'Emission Wavelength (nm)', 'Photoluminescence Quantum Yield',
#        'Photoluminescence Lifetime (ns)', 'Spectral Gain Factor (cm^2 s)',
#        'Emission Gain Cross Section (cm^2)'],
#       dtype='object')

data.nunique()
# Identifier                            1129
# SMILES                                1129
# Synthesis Procedure                      3
# Instrument                               3
# Optical Measurements                     5
# Absorption Wavelength (nm)              99
# Emission Wavelength (nm)               144
# Photoluminescence Quantum Yield        362
# Photoluminescence Lifetime (ns)        362
# Spectral Gain Factor (cm^2 s)            0
# Emission Gain Cross Section (cm^2)     360
# dtype: int64

# plotly scatter matrix
import plotly.express as px
fig = px.scatter_matrix(data,
                        dimensions=["Absorption Wavelength (nm)", "Emission Wavelength (nm)", "Photoluminescence Quantum Yield", "Photoluminescence Lifetime (ns)", "Emission Gain Cross Section (cm^2)"],
                        color="Instrument",)

fig.update_traces(diagonal_visible=False)

fig.update_layout(
    title='ACDC dataset scatter matrix',
    width=1200,
    height=1000,
    font=dict(
        size=6,
    )
)

fig.show()
```
