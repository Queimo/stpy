```python
import pickle
from pathlib import Path
import datetime
import numpy as np
import pandas as pd

# ..\output\2024-11-01\16-14-13_data_dict.pkl

paths_list = list(Path('../output').rglob('*.pkl'))
paths_list

import seaborn as sns
from matplotlib import pyplot as plt


dicts_list = []
for path in paths_list:
    with open(path, 'rb') as f:
        data_dict = pickle.load(f)
        print(data_dict.keys())
        dicts_list.append(data_dict)
        
        df = pd.DataFrame(data_dict) *-1
        # print(df.head())
        plt.figure()
        hist = sns.boxplot(data=df, width=0.5)
        hist.set_title(path)
        # add text n-folds inside the plot area top left
        hist.text(0., 0., f'N-folds: {df.shape[0]}', fontsize=12, ha='left', va='bottom', transform=hist.transAxes)
        
```
