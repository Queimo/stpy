```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
```

```python
df = pd.read_csv('../investigate_cv.csv', index_col=0)
df
```

```python
# cvar_90 of columns
fold_columns = [str(i) for i in range(10)]

df_fold = df[fold_columns]


#where smaller than q90
df["cvar"] = df_fold[df_fold < df['q50'].values[:, None]].mean(axis=1)

# Parameters for the smooth transition
alpha = 10  # Controls the sharpness of the transition (higher = closer to discrete)
weights = 1 / (1 + np.exp(alpha * (df_fold - df['q50'].values[:, None])))
weights /= np.sum(weights.values, axis=1, keepdims=True)

# Compute the weighted mean
df["cvar_smooth"] = (df_fold * weights).sum(axis=1)
df["rel_std"] = df["cv_std"] / df["cv_loss"]

 
```

```python
#find peaks scipy
from scipy.signal import find_peaks
```

```python
plot_cols = ["cvar", "cvar_smooth", "q10", "log_marginal", "cv_loss", "rel_std"]
fig = px.line(df, x="gamma", y=plot_cols, title="CVAR, q90, q50, q10, log_marginal, cv_loss")

# top3 minima of all the plotted columns
#gamma index
for col in plot_cols:
    peaks, _ = find_peaks(-df[col])
    top3 = df.iloc[peaks].sort_values(col).head(3)
    fig.add_scatter(x=top3["gamma"], y=top3[col], mode="markers", marker=dict(size=10), name=col)
    
fig.update_layout(
    xaxis_title="gamma",
    yaxis_title="value",
    legend_title="columns",
    width=1000,
    height=600
)
fig.show()

```
